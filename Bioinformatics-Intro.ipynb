```python
def FrequentWords(Text, k):
    words = []
    freq = FrequencyMap(Text, k)
    m = max(freq.values())
    for key in freq:
        if freq[key] == m:
            words.append(key)
    return words

def FrequencyMap(Text, k):
    freq = {}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        freq[Pattern] = 0
        for i in range(n-k+1):
            if Text[i:i+k] == Pattern:
                freq[Pattern] += 1
    return freq
```

```python
def SymbolArray(Genome, symbol):
    array = {}
    n = len(Genome)
    ExtendedGenome = Genome + Genome[0:n//2]
    for i in range(n):
        array[i] = PatternCount(symbol, ExtendedGenome[i:i+(n//2)])
    return array
```

```python
def FasterSymbolArray(Genome, symbol):
    array = {}
    n = len(Genome)
    ExtendedGenome = Genome + Genome[0:n//2]

    # look at the first half of Genome to compute first array value
    array[0] = PatternCount(symbol, Genome[0:n//2])

    for i in range(1, n):
        # start by setting the current array value equal to the previous array value
        array[i] = array[i-1]

        # the current array value can differ from the previous array value by at most 1
        if ExtendedGenome[i-1] == symbol:
            array[i] = array[i]-1
        if ExtendedGenome[i+(n//2)-1] == symbol:
            array[i] = array[i]+1
    return array
```

```python
def SkewArray(Genome):
    skew = [0]
    score = {'A': 0, 'T': 0, 'C': -1, 'G': 1}
    for i in range(1, len(Genome) + 1):
        skew.append(score[Genome[i-1]] + skew[i-1])
    return skew
```

```python
def MinimumSkew(Genome):
    positions = []
    skew = SkewArray(Genome)
    mini = min(skew)
    for i in range(0, len(skew)):
        if skew[i] == mini:
            positions.append(i)
    return positions
```

```python
def MaxSkew(Genome):
    positions = []
    skew = SkewArray(Genome)
    maxi = max(skew)
    for i in range(0, len(skew)):
        if skew[i] == maxi:
            positions.append(i)
    return positions
```

```python
def HammingDistance(p, q):
    count = 0
    for i, j in zip(p, q):
        if i != j:
            count += 1
    return count
```

```python
def ApproximatePatternMatching(Text, Pattern, d):
    positions = []
    for i in range(len(Text)-len(Pattern) + 1):
        if HammingDistance(Text[i:i+len(Pattern)], Pattern) <= d:
            positions.append(i)
    return positions
```

```python
def ApproximatePatternCount(Text, Pattern, d):
    count = 0
    for i in range(len(Text)-len(Pattern) + 1):
        if HammingDistance(Text[i:i+len(Pattern)], Pattern) <= d:
            count += 1
    return count
```

```python
def Profile(Motifs):
    profile = {}
    k = len(Motifs[0])
    for symbol in "ACGT":
        profile[symbol] = []
        for j in range(k):
            profile[symbol].append(0)
    t = len(Motifs[0])
    for i in range(t):
        for j in range(k):
            symbol = Motifs[i][j]
            profile[symbol[j]] += (1/t)
    return profile
```

```python
def Consensus(Motifs):
    k = len(Motifs[0])
    count = Count(Motifs)
    consensus = ""
    for j in range(k):
        m = 0
        frequentSymbol = ""
        for symbol in "ACGT":
            if count[symbol][j] > m:
                m = count[symbol][j]
                frequentSymbol = symbol
        consensus += frequentSymbol
    return consensus
```

```python
def Score(Motifs):
    consensus = Consensus(Motifs)
    count = 0
    for motif in Motifs:
        for i in range(len(motif)):
            if motif[i] != consensus[i]:
                count += 1
    return count
```

```python
def Pr(Text, Profile):
    p = 1
    for i, b in enumerate(Text):
        p *= Profile[b][i]
    return p
```

```python
def ProfileMostProbableKmer(Text, k, Profile):
    maxi = -1
    x = 0
    for i in range(len(Text)-k+1):
        out = Pr(Text[i:i+k], Profile)
        if out > maxi:
            maxi = out
            x = i
    ans = Text[x:x+k]
    return ans
```

```python
import itertools
from collections import defaultdict

def FrequentWordswithMismatches(Text, k, d):
    words = []
    freq = defaultdict(lambda:0)
    for index in range(len(Text)- k + 1):
        curr_kmer_and_neighbors = PermuteMotifDist(Text[index:index+k], d)
        for kmer in curr_kmer_and_neighbors:
            freq[kmer] += 1
    for kmer in freq:
        if freq[kmer] == max(freq.values()):
            words.append(kmer)
    return words

def PermuteOnce(motif, alphabet={"A", "C", "G", "T"}):
    return list(set(itertools.chain.from_iterable([[
        motif[:pos] + nuc + motif[pos + 1 :] for nuc
        in alphabet] for pos in range(len(motif))])))

def PermuteMotifDist(motif, d):
    workingSet = {motif}
    for _ in range(d):
        workingSet = set(itertools.chain.from_iterable(map(PermuteOnce, workingSet)))
    return list(workingSet)
```

```python
def MotifEnumeration(Dna, k, d):
    Pattern = []
    Patterns2 = {}
    for i in range(0, (len((Dna[0]))-(k+1))):
        neighbors = iterativeneighbors(Dna[0][i:i+k], d)
        for j in neighbors:
            count = 0
            for m in Dna:
                for i in range(0, len(m)-k+1):
                    if HammingDistance(j, m[i:i+k])<=d:
                        count += 1
                        break
            if count == len(Dna):
                Pattern.append(j)
                Patterns2.update({j : count})
    Patterns = []
    [Patterns.append(x) for x in Pattern if x not in Patterns]
    Result = ""
    for item in Patterns:
        Result = Result + " " + str(item)
    return Result, Patterns2
```
